```
import numpy as np
import matplotlib.pyplot as plt

#funcion rand: numeros aleatorios en una distribucion uniforme

print(np.random.rand(3,3))

rand = np.random.rand(20000)
plt.hist(rand, bins=30)
plt.show()

#funcion uniform: igual que rand, pero permite ingresar los l√≠mites de la 
#muestra.

print(np.random.uniform(low=10, high=20, size=6))

rand = np.random.rand(10000)
plt.hist(rand, bins=60)
plt.show()

uniform = np.random.uniform(low=0, high=1, size=10000)
plt.hist(uniform, bins=50)
plt.show()

#funcion randn: numeros aleatorios en una distribucion
print(np.random.randn(2, 2))


#funccion normal: igual que randn pero permite escalar los limites de la prueba

normal = np.random.randn(10000)
plt.hist(normal, bins=50)
plt.show()

print(np.random.normal(loc=0, scale=2, size=6))

normal2 = np.random.normal(0, 1, 10000)
plt.hist(normal2, bins=50)
plt.show()

#funcion randint: numeros aleatorios entre un rango dado
print(np.random.randint(low=0, high=10, size=(3, 3)))
print(np.random.randint(1,100,10))
```
